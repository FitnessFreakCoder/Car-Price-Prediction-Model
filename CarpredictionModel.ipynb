```python
import pandas as pd
import re

# Load dataset


# Clean column names (remove spaces)
data.columns = data.columns.str.strip()

# Function to extract numbers
def extract_number(x):
    """Extract numeric value from strings like '3990 cc', '250 km/h', '900 Nm', '2.5 sec' """
    if pd.isnull(x):
        return None
    x = str(x)
    nums = re.findall(r"[\d\.]+", x.replace(",", ""))  # digits + decimals
    if len(nums) == 0:
        return None
    if "—" in x or "-" in x:   # handle ranges like "12,000—15,000"
        nums = [float(n) for n in nums]
        return sum(nums) / len(nums)   # average
    return float(nums[0])

# Apply numeric cleaning
if "CC/Battery Capacity" in data.columns:
    data["CC"] = data["CC/Battery Capacity"].apply(extract_number)

if "HorsePower" in data.columns:
    data["HorsePower"] = data["HorsePower"].apply(extract_number)

if "Total Speed" in data.columns:
    data["TopSpeed"] = data["Total Speed"].apply(extract_number)

if "Performance(0 - 100 )KM/H" in data.columns:
    data["Acceleration"] = data["Performance(0 - 100 )KM/H"].apply(extract_number)

if "Torque" in data.columns:
    data["Torque"] = data["Torque"].apply(extract_number)

if "Cars Prices" in data.columns:
    data["Price"] = data["Cars Prices"].apply(extract_number)

# Drop old string columns (optional)
cols_to_drop = ["CC/Battery Capacity", "Cars Prices", "Total Speed", "Performance(0 - 100 )KM/H"]
data_clean = data.drop([c for c in cols_to_drop if c in data.columns], axis=1)

# Save cleaned dataset
data_clean.to_csv("cardata_clean.csv", index=False)

print("✅ Numeric columns cleaned! Preview:")
print(data_clean.head())
```
